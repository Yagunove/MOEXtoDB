```python
##final build
#pip install pyodbc

import requests
import pandas as pd
import json
import pyodbc


# Connection variables
server = 'localhost'
database = 'master'
username = 'sa'
password = 'someThingComplicated1234'

#connection string
conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=' +
        server+';DATABASE='+database+';UID='+username+';PWD=' + password)
cursor = conn.cursor()

##API variables
ticker = 'LKOH'
date_from = '2021-07-10'

##making the API call
url = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boards/TQBR/securities/'+ticker+'.json?iss.json=extended&from='+date_from+''

payload={}
headers = {
  'Cookie': 'MPTrack=343fd159.5c888e58c01e2; MicexTrackID=70.79.146.234.1627495388255285'
}

response = requests.request('GET', url, headers=headers, data=payload)

##parsing JSON object into data structure (dict/list)
values = json.loads(response.text)

##in fact it is a list of dictionaries of dictionaries, so we dig deeper
data = values[1]['history'][1]


##preparing df to load it later into DB
df = pd.DataFrame(data)
df = df[['TRADEDATE', 'SECID','CLOSE']]

##the whole nine yards with SQL (create table, buffer, insert, drop buffer)

#create final table
cursor.execute("IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='t_stock_history' AND xtype='U') CREATE TABLE [dbo].[t_stock_history](	[TRADEDATE] date,	[SECID] [nvarchar](255) NULL ,	[CLOSE] float )")
conn.commit()

#create bugger
cursor.execute('CREATE TABLE [dbo].[buff_stock_history](	[TRADEDATE] date,	[SECID] [nvarchar](255) NULL ,	[CLOSE] float )')
conn.commit()

#insert data from API to buffer
for index, row in df.iterrows():
    cursor.execute('INSERT INTO dbo.buff_stock_history ([TRADEDATE],[SECID],[CLOSE]) VALUES(?,?,?)', row.TRADEDATE, row.SECID, row.CLOSE)
conn.commit()

#deduplicating delete-insert. The idea is for a user to request needed ticker for any given time period
#and in case of having duplicate rows already in final table, we will simply update them with this delete-insert technique
cursor.execute('DELETE t1 FROM [dbo].[t_stock_history] t1 JOIN [dbo].[buff_stock_history] t2 ON t1.TRADEDATE = t2.TRADEDATE AND t1.SECID = t2.SECID AND t1.[CLOSE]=t2.[CLOSE]')
conn.commit()

cursor.execute('INSERT INTO [dbo].[t_stock_history] SELECT * FROM [dbo].[buff_stock_history]')
conn.commit()


#drop buffer
cursor.execute('DROP TABLE [dbo].[buff_stock_history]')
conn.commit()


#ending
cursor.close()
del cursor
 
# Close the database connection
conn.close()


```

```python
#### CONNECTION #####
def connection():
    import pyodbc
    import json

    # setting parameters
    with open('connection.json') as f:
        conn_element = json.load(f)

    # pyodbc connection string
    conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=' 
                          +conn_element["server"]
                          +';DATABASE='
                          +conn_element["database"]
                          +';UID='
                          +conn_element["username"]
                          +';PWD=' 
                          + conn_element["password"], autocommit=True)
    
    # initializing cursor
    cursor = conn.cursor()
    
    print("Successfully connected")
    return(conn,cursor)
```

```python
import datetime as DT

### MAKING API CALL TO MOEX
def api_call(ticker = 'LKOH', date_from=0): 
    
    ### If no date is given, take one week ago from current date
    if date_from==0:
        date_from = (DT.datetime.today()-DT.timedelta(days=7)).strftime('%Y-%m-%d')
     
    ### Check validity of the given date
    try:
        DT.datetime.strptime(date_from, '%Y-%m-%d')
    except ValueError:
        raise ValueError("Incorrect data format, should be YYYY-MM-DD")

    ### Making the API call
    url = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boards/TQBR/securities/'+ticker+'.json?iss.json=extended&from='+date_from+''
    response = requests.request('GET', url)
    
    ### Parsing JSON object into data structure (list of dict) and then parsing out values
    values = json.loads(response.text)
    data = values[1]['history'][1]
    
    ### Checking the output
    if(not data):
        raise ValueError("Ticker name or date is incorrect")
        
    ### Loading data into pandas dataframe before transfering it to database
    df = pd.DataFrame(data)
    df = df[['TRADEDATE', 'SECID','CLOSE']]
    
    return(df)
```

```python
def db_insert(cursor, df):

    # creating final table with if-not-exists feature which makes this code universal for the initial run and further ones 
    stmt1 = '''
    IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='t_stock_history' AND xtype='U')
    CREATE TABLE [dbo].[t_stock_history]
    ([TRADEDATE] date,[SECID] [nvarchar](255) NULL ,[CLOSE] float)'''

    # creating buffer which later gives us update option with deduplication
    stmt2= '''
    CREATE TABLE [dbo].[buff_stock_history]([TRADEDATE] date,[SECID] [nvarchar](255) NULL ,[CLOSE] float )'''

    # executing statements
    for st in [stmt1,stmt2]:
        cursor.execute(st)

    # insert data from API to buffer
    for index, row in df.iterrows():
        cursor.execute('INSERT INTO dbo.buff_stock_history ([TRADEDATE],[SECID],[CLOSE]) VALUES(?,?,?)', row.TRADEDATE, row.SECID, row.CLOSE)

    # deleting from final table rows that are present in the buffer
    stmt3 = '''
    DELETE t1 
    FROM [dbo].[t_stock_history] t1 
    JOIN [dbo].[buff_stock_history] t2 
    ON t1.TRADEDATE = t2.TRADEDATE AND t1.SECID = t2.SECID AND t1.[CLOSE]=t2.[CLOSE]'''

    stmt4= '''
    INSERT INTO [dbo].[t_stock_history] 
    SELECT * FROM [dbo].[buff_stock_history]'''

    # drop buffer
    stmt5 = 'DROP TABLE [dbo].[buff_stock_history]'

    for st in [stmt3, stmt4, stmt5]:
        cursor.execute(st)
    
    return(0)
```

```python
def close_connection(conn,cursor):
    cursor.close()
    del cursor
    conn.close()
    print("Connection was closed successfully")
    return(0)
```
